# Import Dataset

```python
import pandas as pd

df = pd.read_csv("../Database/sales_data_sample.csv", sep=",", encoding='Latin-1')
df.head()
```

# Examine Columns

```python
df.info()
```

# Split Dataset

```python
# customers tablosu
customers = df[['CUSTOMERNAME', 'PHONE', 'ADDRESSLINE1', 'ADDRESSLINE2', 'CITY', 'STATE', 'POSTALCODE', 'COUNTRY', 'TERRITORY', 'CONTACTLASTNAME', 'CONTACTFIRSTNAME']].drop_duplicates()
customers['id'] = range(1, len(customers) + 1)

# products tablosu
products = df[['PRODUCTCODE', 'PRODUCTLINE', 'MSRP']].drop_duplicates()

# orders tablosu
orders = df[['ORDERNUMBER', 'ORDERDATE', 'STATUS', 'QTR_ID', 'MONTH_ID', 'YEAR_ID', 'DEALSIZE', 'CUSTOMERNAME']].drop_duplicates()
orders = orders.merge(customers[['CUSTOMERNAME', 'id']], on='CUSTOMERNAME').drop(columns='CUSTOMERNAME')
orders = orders.rename(columns={'id': 'customer_id'})

# order_items tablosu
order_items = df[['ORDERNUMBER', 'PRODUCTCODE', 'QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER', 'SALES']]
```

# Export Tables

```python
# Verileri CSV dosyalarÄ±na kaydetme
customers.to_csv('../Database/customers.csv', index=False)
products.to_csv('../Database/products.csv', index=False)
orders.to_csv('../Database/orders.csv', index=False)
order_items.to_csv('../Database/order_items.csv', index=False)
```
